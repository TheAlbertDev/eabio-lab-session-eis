
### 4.2 Fitting del model de imped√†ncia electroqu√≠mica

A continuaci√≥, pugem el fitxer exportat executant el block de codi de seg√ºent i seleccionant el fitxer exportat. Si escollim m√©s d'un fitxer o no √©s un fitxer v√†lid, es notificar√† un error i haurem de repetir el proc√©s de importaci√≥.

```python
from google.colab import files
import os
import pandas as pd
import re

uploaded = files.upload()

if len(uploaded) != 1:
    raise ValueError("Si us plau, puja nom√©s un fitxer.")

filename = list(uploaded.keys())[0]
print("Archivo recibido:", filename)

filename = list(uploaded.keys())[0]

ext = os.path.splitext(filename)[1].lower()

if ext not in ['.csv', '.txt']:
    raise ValueError("Nom√©s es permeten fitxers CSV o TXT.")

print("Fitxer v√†lid:", filename)

try:
    try:
        df = pd.read_csv(filename, sep='\t')
    except:
        df = pd.read_csv(filename, sep=',')
    
    if df.shape[1] != 2:
        raise ValueError(f"El fitxer ha de tenir exactament 2 columnes, per√≤ en t√© {df.shape[1]}.")
    
    expected_cols = ["Freq.", "(V(vref)-V(vwork))/I(Rtia)"]
    if list(df.columns) != expected_cols:
        raise ValueError(f"Els noms de les columnes no s√≥n correctes.\nEsperat: {expected_cols}\nTrobat: {list(df.columns)}")
    
    try:
        pd.to_numeric(df.iloc[:, 0])
    except:
        raise ValueError("La columna de freq√º√®ncia ha de contenir nom√©s valors num√®rics.")
    
    impedance_col = df.iloc[:, 1]
    complex_pattern = r'^-?\d+\.?\d*([eE][+-]?\d+)?,-?\d+\.?\d*([eE][+-]?\d+)?$'
    
    for i, value in enumerate(impedance_col):
        if not isinstance(value, str) or not re.match(complex_pattern, value):
            raise ValueError(f"El format de la imped√†ncia a la fila {i+2} no √©s correcte.\nEsperat: 'real,imaginari' (exemple: '1.23e+05,-4.56e+04')\nTrobat: '{value}'")
    
    print("‚úì Format del fitxer validat correctament!")
    print(f"‚úì Nombre de punts de freq√º√®ncia: {len(df)}")
    print(f"‚úì Rang de freq√º√®ncies: {df.iloc[0, 0]} Hz - {df.iloc[-1, 0]} Hz")
    
except Exception as e:
    raise ValueError(f"Error en validar el format del fitxer: {str(e)}")

```

Un cop importat el fitxer amb √®xit, executa el seg√ºent bloc de codi per visualitzar les dades.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

frequencies = df.iloc[:, 0].astype(float)
impedance_strings = df.iloc[:, 1]

real_parts = []
imag_parts = []

for imp_str in impedance_strings:
    real_str, imag_str = imp_str.split(',')
    real_parts.append(float(real_str))
    imag_parts.append(float(imag_str))

real_parts = np.array(real_parts)
imag_parts = np.array(imag_parts)

impedances = real_parts + 1j * imag_parts

magnitude = np.abs(impedances)
phase_rad = np.angle(impedances)
phase_deg = np.degrees(phase_rad)

fig = plt.figure(figsize=(15, 10))
gs = GridSpec(2, 2, width_ratios=[1.2, 1], height_ratios=[1, 1], hspace=0.3, wspace=0.3)

ax_nyquist = fig.add_subplot(gs[:, 0])
ax_nyquist.plot(real_parts, -imag_parts, 'bo-', markersize=4, linewidth=1.5, label='Dades experimentals')
ax_nyquist.set_xlabel('Z\' (Re) [Œ©]', fontsize=12)
ax_nyquist.set_ylabel('-Z\'\' (-Im) [Œ©]', fontsize=12)
ax_nyquist.set_title('A) Diagrama de Nyquist', fontsize=14, fontweight='bold')
ax_nyquist.grid(True, alpha=0.3)
ax_nyquist.set_aspect('equal', adjustable='box')
ax_nyquist.legend()

freq_labels = [0, len(frequencies)//4, len(frequencies)//2, 3*len(frequencies)//4, -1]
for i in freq_labels:
    ax_nyquist.annotate(f'{frequencies.iloc[i]:.1e} Hz', 
                       (real_parts[i], -imag_parts[i]),
                       xytext=(5, 5), textcoords='offset points',
                       fontsize=8, alpha=0.7)

ax_mag = fig.add_subplot(gs[0, 1])
magnitude_db = 20 * np.log10(magnitude)
ax_mag.semilogx(frequencies, magnitude_db, 'r-', linewidth=2)
ax_mag.set_xlabel('Freq√º√®ncia [Hz]', fontsize=12)
ax_mag.set_ylabel('|Z| [dB]', fontsize=12)
ax_mag.set_title('B) Diagrama de Bode - Magnitud', fontsize=14, fontweight='bold')
ax_mag.grid(True, alpha=0.3)

ax_phase = fig.add_subplot(gs[1, 1])
ax_phase.semilogx(frequencies, phase_deg, 'g-', linewidth=2)
ax_phase.set_xlabel('Freq√º√®ncia [Hz]', fontsize=12)
ax_phase.set_ylabel('Fase [¬∞]', fontsize=12)
ax_phase.set_title('C) Diagrama de Bode - Fase', fontsize=14, fontweight='bold')
ax_phase.grid(True, alpha=0.3)

plt.suptitle('An√†lisi d\'Imped√†ncia Electroqu√≠mica', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nüìä Resum de les dades:")
print(f"Imped√†ncia m√†xima: {np.max(magnitude):.2e} Œ© @ {frequencies.iloc[np.argmax(magnitude)]:.1f} Hz")
print(f"Imped√†ncia m√≠nima: {np.min(magnitude):.2e} Œ© @ {frequencies.iloc[np.argmin(magnitude)]:.1f} Hz")
print(f"Fase m√†xima: {np.max(phase_deg):.1f}¬∞ @ {frequencies.iloc[np.argmax(phase_deg)]:.1f} Hz")
print(f"Fase m√≠nima: {np.min(phase_deg):.1f}¬∞ @ {frequencies.iloc[np.argmin(phase_deg)]:.1f} Hz")
```

Per obtenir els valors de $R_2$ i $C_1$ mitjan√ßant un fitting del nostre model, utilitzarem el paquet de Python [impedance.py](https://impedancepy.readthedocs.io/en/latest/index.html). El paquet necesita que definim quin tipus de cel¬∑la volem fer el _fitting_. S'indica amb una cadena de text on els elements en serie es separen per `-` i els elements en paralel s'envolten amb `p( , )`. En el nostre cas, el string a fer servir ser√≠a `R1-p(R2,C2)`. A m√©s, necessitem indicar uns valors inicials a partir del qual iniciar√° la iteraci√≥ matem√°ticament per trobar els valors finals. En aquest cas es tracta d'una simul¬∑laci√≥ en la que sabems els valors exactes. Amb objectiu academic, evitarem utilitzar els valors exactes i indicarem uns valors propers als utilitzs. Els valors s'indiquen mitjan√ßant un vector de valors on l'ordre dels elements segueix el mateix ordre dels components indicats en el model. En el nostre cas, podriem fer servir `[0.980e6, 1.1e6, 400e-12]`. Executem el seg√ºent block de codi per veure el resultat del _fitting_.

```python
import matplotlib.pyplot as plt
from impedance.models.circuits import CustomCircuit
from impedance.visualization import plot_nyquist

circuit_model = 'R1-p(R2,C2)'
initial_guess = [0.980e6, 1.1e6, 400e-12]
circuit = CustomCircuit(circuit_model, initial_guess=initial_guess)

circuit.fit(frequencies, impedances)

print(circuit)

impedance_fit = circuit.predict(frequencies)

fig, ax = plt.subplots()
plot_nyquist(impedances, fmt='o', scale=10, ax=ax)
plot_nyquist(impedance_fit, fmt='-', scale=10, ax=ax)

plt.legend(['Data', 'Fit'])
plt.show()
```

resist√®ncia $R_1$ no participa en el model donat que, gracies a l‚Äôamplificador operacional U1, la caiguda a la resist√®ncia $R_1$ es negligeix. Indica el model i els valors inicials en el seg√ºent formulari i executa el bloc de codi.

```python
#@title Par√†metres d'entrada pel _fitting_ { display-mode: "form" }
#@markdown Indica el model equivalent que vols ajustar i els valors inicials per als par√†metres.

#@markdown Introdueix el model del teu circuit. 
#@markdown Exemple: `R1-p(R2,C2)`
circuit_model = ''  #@param {type: "string"}

#@markdown Introdueix els valors esperats separats per comes i fent servir el punt com a separador decimal. 
#@markdown Exemple: `10.0, 20, 30.0`
input_initial_guess = ''  #@param {type: "string"}

initial_guess = [float(x.strip()) for x in input_initial_guess.split(',') if x.strip()]

circuit = CustomCircuit(circuit_model, initial_guess=initial_guess)

circuit.fit(frequencies, impedances)

print(circuit)

impedance_fit = circuit.predict(frequencies)

fig, ax = plt.subplots()
plot_nyquist(impedances, fmt='o', scale=10, ax=ax)
plot_nyquist(impedance_fit, fmt='-', scale=10, ax=ax)

plt.legend(['Data', 'Fit'])
plt.show()
```
